## Gaussian vectors

```python
rndn = np.random.randn(500,2)
fig, ax = plt.subplots()
ax.scatter(rndn[:,0],rndn[:,1]) 
plt.show()
```

### Add correlation

```python
mu = [0, 0]#mean
gamma = [[2, 1], [1, 4]]  # diagonal covariance
x, y = np.random.multivariate_normal(mu, gamma, 5000).T  #Generate Gaussian vectors with mean mu and covariance gamma 
```

```python
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```

# Visualization with Python

2D Graphics

We generate data

```python
sequence_containing_x_vals = list(range(0, 100))
sequence_containing_y_vals = list(range(0, 100))
import random
random.shuffle(sequence_containing_x_vals)
random.shuffle(sequence_containing_y_vals)
print([sequence_containing_x_vals,sequence_containing_y_vals])
```

```python
fig = plt.figure()

ax = fig.add_subplot(111)
ax.scatter(sequence_containing_x_vals, sequence_containing_y_vals)
plt.show()
```

### We dilate and rotate

```python
s1 = np.array([[10,0],[0,1]]) #definition of a 2D dilation
```

```python
r1=np.array([[np.sqrt(3)/2,-1/2],[1/2,np.sqrt(3)/2]])#definition of a rotation
```

```python
for i in range(0, 100):
    v=np.array([sequence_containing_x_vals[i],sequence_containing_y_vals[i]]).T
```

```python
print(v)
```

```python
obs=np.zeros((2,100))
for i in range(0, 100):
    v=np.array([sequence_containing_x_vals[i],sequence_containing_y_vals[i]]).T
    obs[:,i]=s1.dot(r1).dot(v)
x_vals=obs[0,:]
y_vals=obs[1,:]
```

```python
fig = plt.figure()

ax = fig.add_subplot(111)
ax.scatter(x_vals, y_vals)
plt.show()
```

3D Graphics

We generate data

```python
sequence_containing_x_vals = list(range(0, 100))
sequence_containing_y_vals = list(range(0, 100))
sequence_containing_z_vals = list(range(0, 100))

import random
random.shuffle(sequence_containing_x_vals)
random.shuffle(sequence_containing_y_vals)
random.shuffle(sequence_containing_z_vals)
```

```python
print([sequence_containing_x_vals,sequence_containing_y_vals,sequence_containing_z_vals])
```

```python
from mpl_toolkits.mplot3d import Axes3D  

import matplotlib.pyplot as plt
import numpy as np


fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')

ax.scatter(sequence_containing_x_vals, sequence_containing_y_vals, sequence_containing_z_vals)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')


plt.show()
```
