# Centering data

```python
n_samples, n_features = faces.shape
print("Dataset consists of %d faces" % n_samples)
```

```python
# global centering of the dataset
faces_centered = faces - faces.mean(axis=0)
```

```python
 # we center now each image 
faces_centered = faces_centered-faces_centered.mean(axis=1).reshape(n_samples, -1)
```

# Visualize the data

We fix some parameters

```python
n_row, n_col = 2, 5
n_components = n_row * n_col#equals 10
image_shape = (64, 64)
print(64*64)
```

We define  a function allowing to plot the dataset

```python
def plot_gallery(title, images, n_col=n_col, n_row=n_row, cmap=plt.cm.gray):
    plt.figure(figsize=(2. * n_col, 2.26 * n_row))
    plt.suptitle(title, size=16)
    for i, comp in enumerate(images):
        plt.subplot(n_row, n_col, i + 1)
        vmax = max(comp.max(), -comp.min())
        plt.imshow(comp.reshape(image_shape), cmap=cmap,
                   interpolation='nearest',
                   vmin=-vmax, vmax=vmax)
        plt.xticks(())
        plt.yticks(())
    plt.subplots_adjust(0.01, 0.05, 0.99, 0.93, 0.04, 0.)
```

```python
plot_gallery("First Olivetti faces", faces[:n_components])
```

```python
plot_gallery("First centered Olivetti faces", faces_centered[:n_components])
```

# PCA

```python
from sklearn.decomposition import PCA# we import the appropriate library
pca = PCA(n_components=3)#we define the model
```

```python
pca.fit(faces)
components_=pca.components_
n_samples, n_features = components_.shape
print(n_samples, n_features)
```

```python
plot_gallery('Composantes principales du dataset',components_)
```

```python
pca.fit(faces_centered)
components_centered_=pca.components_
n_samples, n_features = components_centered_.shape
print(n_samples, n_features)
```

```python
plot_gallery('Composantes principales du dataset',components_centered_)
```

# NMF

```python
from sklearn.decomposition import NMF# we import the appropriate library
nmf = NMF(n_components=3)#we define the model
```

```python
nmf.fit(faces)
components_nmf_=nmf.components_
n_samples, n_features = components_nmf_.shape
print(n_samples, n_features)
```

```python
plot_gallery('Composantes NMF du dataset',components_nmf_)
```

# Question 2

Keep only unique values of movieId and userId

```python
import numpy as np
unique_movie = ratings["movieId"].unique()
unique_id = ratings["userId"].unique()
```

We count how many unique values we have

```python
colonnes = len(unique_movie)
lignes = len(unique_id)
grid = np.zeros((lignes,colonnes)) 
```

We assign in grid unique values

```python
for i in range(ratings.shape[0]):
    grid[ratings["userId"][i]==unique_id,ratings["movieId"][i]==unique_movie]=ratings["rating"][i]
```

# Question 3

Compute NMF of this matrix

```python
import sklearn
from sklearn.decomposition import NMF
model = NMF(n_components=20, init='random', random_state=0)
W = model.fit_transform(grid)
H = model.components_
```

# Question 4

We can now use NMF for prediction

```python
all_user_predicted_ratings = np.dot(W, H)
```

We truncate the value of this matrix 

```python
all_user_predicted_ratings[all_user_predicted_ratings < 0.5] = 0.5
all_user_predicted_ratings[all_user_predicted_ratings > 5] = 5
```

Display predicted values

```python
preds_df = pd.DataFrame(all_user_predicted_ratings, columns = unique_movie, index=unique_id)
preds_df.head()
```

# Question 5

```python
def recommend_movies(predictions_df, userID, movies_df, original_ratings_df, num_recommendations=5):
    
    # Get and sort the user's predictions
    user_row_number = userID - 1 # UserID starts at 1, not 0
    sorted_user_predictions = predictions_df.iloc[user_row_number].sort_values(ascending=False)
    
    # Get the user's data and merge in the movie information.
    user_data = original_ratings_df[original_ratings_df.userId == (userID)]
    user_full = (user_data.merge(movies_df, how = 'left', left_on = 'movieId', right_on = 'movieId').
                     sort_values(['rating'], ascending=False))

    # Recommend the highest predicted rating movies that the user hasn't seen yet.
    SUP=pd.DataFrame(sorted_user_predictions).reset_index()
    SUP.columns=['movieId',user_row_number]
    
    recommendations = (movies_df[~movies_df['movieId'].isin(user_full['movieId'])].
    merge(SUP, how = 'left',left_on = 'movieId',right_on = 'movieId').
    rename(columns = {user_row_number: 'Predictions'}).
    sort_values('Predictions', ascending = False).
    iloc[:num_recommendations, :-1])

    return user_full, recommendations
```

```python
already_rated, predictions = recommend_movies(preds_df, 0, movies, ratings, 1)
predictions
```
