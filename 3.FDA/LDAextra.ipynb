# We work on a dataset from the UCI ML repository

```python
import pandas as pd
```

```python
columns =["var","skewness","kurtosis","entropy","class"]
```

```python
df = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/00267/data_banknote_authentication.txt",index_col=False,names = columns)
df
```

```python
df["var"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
f, ax = plt.subplots(1, 4, figsize=(10,3))
vis1 = sns.distplot(df["var"],bins=10, ax= ax[0])
vis2 = sns.distplot(df["skewness"],bins=10, ax=ax[1])
vis3 = sns.distplot(df["kurtosis"],bins=10, ax=
ax[2])
vis4 = sns.distplot(df["entropy"],bins=10, ax=ax[3])
```

```python
df_values=df.values
df_values
```

```python
df_data=df_values[:,:4]
df_label=df_values[:,4]
```

```python
import sklearn
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
lda.fit(df_data,df_label)
```

```python
df_data_trans = lda.transform(df_data)
plt.plot(df_data_trans, df_label, 'r+')
plt.show()
```
